# Einstein Summation Convetion Method

This function is used to compute many multi-dimnensional and linear algebric operations on arrays.

```python
a = np.arange(1,10).reshape(3,3)
b = np.arange(11,20).reshape(3,3)
a,b
```

```python
## Extracts diagonals
np.einsum('ii->i',a)
```

```python
## Transpose
np.einsum('ij->ji',a)
```

```python
a.T
```

```python
## Determinant
print(np.linalg.det(a))
```

```python
np.linalg.inv(b)
```

```python
## Matrix Multiplication
np.einsum('ij,jk',a,b)
```
